# Forecast

```python
ts_train, ts_test = df_handle.forecast_setup(stationary_ts, 6500)
```

## \Simple Exponential Smoothing\

```python
smoothing_level = 0

forecaster = modeling.SimpleExpSmoothing(ts_train).fit(smoothing_level=smoothing_level, optimized=True)
```

```python
model_evaluation.test_forecast(forecaster=forecaster, ts_test=ts_test, ts_train=ts_train, ts=stationary_ts)
```

```python
smoothing_level=0.1
smoothing_slope=0.2

forecaster = modeling.Holt(ts_train).fit(smoothing_level=smoothing_level, smoothing_slope=smoothing_slope)
```

```python
model_evaluation.test_forecast(forecaster=forecaster, ts_test=ts_test, ts_train=ts_train, ts=stationary_ts)
```

```python
seasonal_periods=1000

forecaster = modeling.ExponentialSmoothing(ts_train, seasonal_periods=seasonal_periods, trend=None, seasonal='add').fit()
```

```python
model_evaluation.test_forecast(forecaster=forecaster, ts_test=ts_test, ts_train=ts_train, ts=stationary_ts)
```

```python
forecaster = modeling.ARIMA(ts_train, order=(3,0,1)).fit()
```

```python
model_evaluation.test_forecast(forecaster=forecaster, ts_test=ts_test, ts_train=ts_train, ts=stationary_ts)
```

```python
forecaster = modeling.SARIMAX(ts_train, order=(1,0,0), seasonal_order=(10, 1, 0, 1)).fit(disp=False)
```

```python
model_evaluation.test_forecast(forecaster=forecaster, ts_test=ts_test, ts_train=ts_train, ts=stationary_ts)
```
